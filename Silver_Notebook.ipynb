##### **Read File From Lakehouse**

```python
df_churn = spark.read.csv("Files/Bronze/churn.csv", header=True, inferSchema=True)
df_customers = spark.read.csv("Files/Bronze/customers.csv", header=True, inferSchema=True)
df_orders = spark.read.csv("Files/Bronze/orders.csv", header=True, inferSchema=True)
df_products = spark.read.csv("Files/Bronze/products.csv", header=True, inferSchema=True)
df_sales = spark.read.csv("Files/Bronze/sales.csv", header=True, inferSchema=True)
```

##### **Load to silver schema**

```python
df_churn.write.format("delta").saveAsTable("silver.churn")
df_customers.write.format("delta").saveAsTable("silver.customers")
df_orders.write.format("delta").saveAsTable("silver.orders")
df_products.write.format("delta").saveAsTable("silver.products")
df_sales.write.format("delta").saveAsTable("silver.sales")
```
